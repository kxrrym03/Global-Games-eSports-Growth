```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/global_gaming_esports_2010_2025.csv")
display(df.head())
```

```python
sns.countplot(data=df, x='Region')
plt.title('Distribution of Regions')
plt.xlabel('Region')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df, x='Top_Platform')
plt.title('Distribution of Platforms')
plt.xlabel('Platform')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df, x='Top_Genre')
plt.title('Distribution of Genres')
plt.xlabel('Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
country_performance = df.groupby('Country')[['Gaming_Revenue_BillionUSD', 'Esports_Revenue_MillionUSD', 'Active_Players_Million']].mean()
display(country_performance.head())
```

```python
least_gaming_revenue_countries = country_performance.sort_values(by='Gaming_Revenue_BillionUSD', ascending=True)
best_gaming_revenue_countries = country_performance.sort_values(by='Gaming_Revenue_BillionUSD', ascending=False)

print("Least Gaming Revenue Countries:")
display(least_gaming_revenue_countries.head())

print("\nBest Gaming Revenue Countries:")
display(best_gaming_revenue_countries.head())
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=best_gaming_revenue_countries.head(10).index, y=best_gaming_revenue_countries.head(10)['Gaming_Revenue_BillionUSD'])
plt.title('Top 10 Countries by Average Gaming Revenue')
plt.xlabel('Country')
plt.ylabel('Average Gaming Revenue (Billion USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
least_esports_revenue_countries = country_performance.sort_values(by='Esports_Revenue_MillionUSD', ascending=True)
best_esports_revenue_countries = country_performance.sort_values(by='Esports_Revenue_MillionUSD', ascending=False)

print("Least Esports Revenue Countries:")
display(least_esports_revenue_countries.head())

print("\nBest Esports Revenue Countries:")
display(best_esports_revenue_countries.head())
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=best_esports_revenue_countries.head(10).index, y=best_esports_revenue_countries.head(10)['Esports_Revenue_MillionUSD'])
plt.title('Top 10 Countries by Average Esports Revenue')
plt.xlabel('Country')
plt.ylabel('Average Esports Revenue (Million USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
least_active_players_countries = country_performance.sort_values(by='Active_Players_Million', ascending=True)
best_active_players_countries = country_performance.sort_values(by='Active_Players_Million', ascending=False)

print("Least Active Players Countries:")
display(least_active_players_countries.head())

print("\nBest Active Players Countries:")
display(best_active_players_countries.head())
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=best_active_players_countries.head(10).index, y=best_active_players_countries.head(10)['Active_Players_Million'])
plt.title('Top 10 Countries by Average Active Players')
plt.xlabel('Country')
plt.ylabel('Average Active Players (Million)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
print("\n--- Summary of Findings ---\n")

print("Top and Bottom Country by Average Gaming Revenue:")
print(f"Best: {best_gaming_revenue_countries.index[0]} with {best_gaming_revenue_countries['Gaming_Revenue_BillionUSD'].iloc[0]:.2f} Billion USD")
print(f"Least: {least_gaming_revenue_countries.index[0]} with {least_gaming_revenue_countries['Gaming_Revenue_BillionUSD'].iloc[0]:.2f} Billion USD")

print("\nTop and Bottom Country by Average Esports Revenue:")
print(f"Best: {best_esports_revenue_countries.index[0]} with {best_esports_revenue_countries['Esports_Revenue_MillionUSD'].iloc[0]:.2f} Million USD")
print(f"Least: {least_esports_revenue_countries.index[0]} with {least_esports_revenue_countries['Esports_Revenue_MillionUSD'].iloc[0]:.2f} Million USD")

print("\nTop and Bottom Country by Average Active Players:")
print(f"Best: {best_active_players_countries.index[0]} with {best_active_players_countries['Active_Players_Million'].iloc[0]:.2f} Million Players")
print(f"Least: {least_active_players_countries.index[0]} with {least_active_players_countries['Active_Players_Million'].iloc[0]:.2f} Million Players")
```
